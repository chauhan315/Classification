```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('HR-Employee-Attrition.csv')
```

```python
data.head()
```

```python
data.info()
```

###### deleting EmployeeCount and StandardHours columns, as they are same for each row, and therefore has no real influence on the target field

```python
data.drop(columns=['EmployeeCount', 'StandardHours'], inplace=True)
```

```python
data.head()
```

###### changing Yes/No to binary in the Attrition column

```python
data.Attrition.replace({"Yes":1, "No":0}, inplace=True)
data.tail()
```

###### now we encode the categorical values

```python
cat_cols = data.select_dtypes(exclude=np.number).columns
num_cols = data.select_dtypes(include=np.number).columns
```

```python
cat_cols
```

```python
num_cols
```

###### get the dummies and merge the dataframe

```python
temp = pd.get_dummies(data[cat_cols], drop_first=True, dtype='int64')
final_data = pd.concat([data, temp], axis=1)
```

```python
final_data.drop(columns=cat_cols, inplace=True, axis=1)
```

```python
final_data.head()
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
```

###### train and test spliting of the data

```python
X = final_data.drop(columns='Attrition', inplace=False)
y = final_data.pop('Attrition')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=10)
```

 ### create Gaussian Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
```

###### train and predict using training data set

```python
model.fit(X_train, y_train)
```

```python
pred1 = model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix

con_mat = confusion_matrix(y_pred=pred1, y_true=y_train)
con_mat
```

```python
training_accuracy_percent = metrics.accuracy_score(y_true=y_train, y_pred=pred1)
training_accuracy_percent
```

##### our model's accuracy is 82% from the training set

###### predict using test data set

```python
test_pred = model.predict(X_test)
```

```python
con_mat2 = confusion_matrix(y_test, test_pred)
con_mat2
```

```python
test_accuracy_percent = metrics.accuracy_score(y_test, test_pred)
test_accuracy_percent
```
